```python
data.drop(['num_outbound_cmds','is_host_login','num_compromised','su_attempted','num_file_creations','is_guest_login','dst_host_rerror_rate'],axis=1)
test.drop(['num_outbound_cmds','is_host_login','num_compromised','su_attempted','num_file_creations','is_guest_login','dst_host_rerror_rate'],axis=1)
```

```python
data.loc[data['label'] != "normal." , 'label']= 1
data.loc[data['label'] == "normal." , 'label']= 0
test.loc[test['label'] != "normal." , 'label']= 1
test.loc[test['label'] == "normal." , 'label']= 0
```

```python
data['label'] = data['label'].astype(float)
test['label'] = test['label'].astype(float)
```

```python
from sklearn.feature_extraction import DictVectorizer
categorical_features = data[['protocol_type', 'service', 'flag']]
dv = DictVectorizer()
cat_matrix = dv.fit_transform(categorical_features.T.to_dict().values())
cat_matrix
```

```python
from scipy.sparse import hstack
other_features = data[['src_bytes', 'dst_bytes', 'num_failed_logins', 'logged_in', 'num_compromised', 'root_shell', 'su_attempted', 'num_root', 'num_file_creations', 'num_shells', 'num_access_files', 'num_outbound_cmds']]
data_matrix = hstack([cat_matrix, other_features])
data_matrix
```

```python
from sklearn.preprocessing import scale
data_matrix = scale(data_matrix.todense())
data_matrix
```

```python
data2 = data_matrix
```

```python
from sklearn.decomposition import PCA
from sklearn import cluster
from scipy.cluster.vq import kmeans,vq
import pylab 

reduced_data = PCA(n_components=2).fit_transform(data2)
centroids,_ = kmeans(reduced_data,2)
idx,_ = vq(reduced_data,centroids)
```

```python
import pylab



# some plotting using numpy's logical indexing
pylab.ylim([-100,10])
pylab.xlim([-2,6])
pylab.plot(reduced_data[idx==0,0],reduced_data[idx==0,1],'or',
     reduced_data[idx==1,0],reduced_data[idx==1,1],'om')
pylab.plot(centroids[:,0],centroids[:,1],'sg',markersize=8)
pylab.show()

```
