```python
import numpy as np
import pandas as pd
col_names = ["duration","protocol_type","service","flag","src_bytes",
    "dst_bytes","land","wrong_fragment","urgent","hot","num_failed_logins",
    "logged_in","num_compromised","root_shell","su_attempted","num_root",
    "num_file_creations","num_shells","num_access_files","num_outbound_cmds",
    "is_host_login","is_guest_login","count","srv_count","serror_rate",
    "srv_serror_rate","rerror_rate","srv_rerror_rate","same_srv_rate",
    "diff_srv_rate","srv_diff_host_rate","dst_host_count","dst_host_srv_count",
    "dst_host_same_srv_rate","dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
    "dst_host_rerror_rate","dst_host_srv_rerror_rate","label"]
data = pd.read_csv('C:/Users/hp lap/Desktop/CS461/Project/dataset/train_data.csv', header=None, names = col_names)
```

```python
# these features are not important. removing these will not affect the result
```

```python
data.drop(['num_outbound_cmds','is_host_login','num_compromised','su_attempted','num_file_creations','is_guest_login','dst_host_rerror_rate'],axis=1)
```

```python
discrete_col=data.select_dtypes(exclude=[object]).columns.values
```

```python
continous_col=(['protocol_type', 'service', 'flag'])
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for cols in continous_col:
    data[str(cols)] = le.fit_transform(data[cols]) 
```

```python
data.loc[data['label'] != "normal." , 'label']= 1
data.loc[data['label'] == "normal." , 'label']= 0
```

```python
data['label'] = data['label'].astype(float)
```

```python
test = pd.read_csv('C:/Users/hp lap/Desktop/CS461/Project/dataset/test_data.csv', header=None, names = col_names)
test.head()
```

```python
test.drop(['num_outbound_cmds','is_host_login','num_compromised','su_attempted','num_file_creations','is_guest_login','dst_host_rerror_rate'],axis=1)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for cols in continous_col:
    test[str(cols)] = le.fit_transform(test[cols]) 
```

```python
test.loc[test['label'] != "normal." , 'label']= 1
test.loc[test['label'] == "normal." , 'label']= 0
test.head()
```

```python
test['label'] = test['label'].astype(float)
```

```python
training = np.array(data)

da_train_data_length = len(data.columns)-2
train_data = training[:,0:da_train_data_length]

da_train_result_length = len(data.columns)-1
train_result = training[:,[da_train_result_length]]
```

```python
testing = np.array(test)

da_test_data_length = len(test.columns)-2
test_data = testing[:,0:da_test_data_length]

da_test_result_length = len(test.columns)-1
test_result = testing[:,[da_test_result_length]]
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

clf = GaussianNB()
clf.fit(train_data,train_result.ravel())
pred_result = clf.predict(test_data)
metrics.accuracy_score(test_result,pred_result)
```

```python
from sklearn import linear_model
from sklearn import metrics

clf = linear_model.LogisticRegression()
clf.fit(train_data,train_result.ravel())
pred_result = clf.predict(test_data)
metrics.accuracy_score(test_result,pred_result)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

clf = RandomForestClassifier()
clf.fit(train_data,train_result.ravel())
pred_result = clf.predict(test_data)
metrics.accuracy_score(test_result,pred_result)
```
